```python
from math import *
from SimpleGraphics import *
```

```python
# Set constants
xcs = 0
ycs = 0
xce = 1200
yce = 900
dx = 30
dy = 30
ticksize = 4
x_label_offset = 12
y_label_offset = 16
step = 1

# Determine origin position
ox = (xce-xcs)/2
oy = (yce-ycs)/2

# Determine minimum and maximum x- and y-axis limits
xmin = -int(ox/dx)
xmax = -xmin+1
ymin = -int(oy/dy)
ymax = -ymin+1

# Change the background white
setOutline("white")
background("white")
```

```python
# Draw the Axes

# Draw lines for axes
setOutline("black")
line(xcs,oy,xce,oy)
line(ox,ycs,ox,yce)

# Draw x-tick marks and x-labels
for i in range(xmin,xmax):
    xval = ox+i*dx
    ystart = oy-ticksize
    yend = oy+ticksize
    line(xval,ystart,xval,yend)
    xlabel = str(i)
    ytxt = oy+x_label_offset
    if i!=0:
        text(xval,ytxt,xlabel)
        
# Draw y-tick marks and y-labels
for i in range(ymin,ymax):
    xstart = ox-ticksize
    xend = ox+ticksize
    yval = oy+i*dy
    line(xstart,yval,xend,yval)
    ylabel = str(-i)
    xtxt = ox+y_label_offset
    if i!=0:
        text(xtxt,yval,ylabel)
```

```python
# Create variables to monitor whether blank line has been entered, and the number of curves plotted
looping = True
count = 0

# Loop until a blank line is entered
while (looping):
    expression = input("Enter an arithmetic expression (blank line to quit): ")
    if (expression == ""):
        looping = False

    # Calculate and set the color to use for the curve
    color = count % 3
    if (color == 0):
        col = "red"
    elif (color == 1):
        col = "green"
    elif (color == 2):
        col = "blue"
    setFill(col)
    setOutline(col)
    
    # If a blank line was NOT entered, plot the curve, as well as local maximum and minimum points
    
    xmm = []
    ymm = []
    mmcol = []
    
    if (looping):
        points = []
        compare = -10000
        psi = -10000
        psi2 = -10000
        for xpixel in range(xcs,xce,step):
            # Get the (x,y) positions of the start of the line segment
            x = (xpixel-ox)/dx
            xstart = x*dx+ox
            y = eval(expression)
            ystart = -y*dy+oy
            # Get the (x,y) positions of the end of the line segment
            x = (xpixel+step-ox)/dx
            xend = x*dx+ox
            y = eval(expression)
            yend = -y*dy+oy
            # Store the (x,y) position values of the start of the line segment in list for plotting later
            points.append(xstart)
            points.append(ystart)
            
            # From the second xpixel position onwards, decide whether the curve is increasing or decreasing.
            # Compare to the previous step.  If we go from increasing to decreasing, or vice versa, that is 
            # the indicator of either a maximum or minimum.  Store the (x,y) position of the max/min for
            # plotting later, as well as the appropriate color.
            if (xpixel > xcs):
                if (yend>ystart):
                    compare = 1
                elif (yend==ystart):
                    compare = 0
                else:
                    compare = -1
                if (compare != psi):
                    #print ("current max/min at x = ",xstart,ystart,xend,yend,yend-ystart)
                    #print ("prev max/min at x = ",xstart_prev,ystart_prev,xend_prev,yend_prev,yend_prev-ystart_prev)
                    if (compare == 1):
                        if (psi == -1 or (psi == 0 and psi2 == -1)):
                            mmcol.append("purple")
                            xmm.append(xstart)
                            ymm.append(ystart)
                    if (compare == -1):
                        if (psi == 1 or (psi == 0 and psi2 == 1)):
                            mmcol.append("orange")
                            xmm.append(xstart)
                            ymm.append(ystart)

                    
            # Remember whether the curve is increasing or decreasing, for comparision purposes at the
            # next step.
            psi2 = psi
            if (yend>ystart):
                psi = 1
            elif (yend == ystart):
                psi = 0
            else:
                psi = -1
            xstart_prev = xstart
            ystart_prev = ystart
            xend_prev = xend
            yend_prev = yend
        
        # Record the last (x,y) point of the curve for plotting.
        points.append(xend)
        points.append(yend)
        # Plot the curve
        line(points)    

        # Plot the maximum/minimum points
        for i in range(len(xmm)):
            setFill(mmcol[i])
            setOutline(mmcol[i])
            ellipse(xmm[i]-ticksize,ymm[i]-ticksize,2*ticksize,2*ticksize)
        setFill(col)
        setOutline(col)
    
    # Increment the number of plotted curves
    count = count + 1

# Close the graphics window
close()
    
```
