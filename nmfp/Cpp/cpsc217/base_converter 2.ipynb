```python
#
# convert any number from decimal to a given base
#
def base10toN(num, base):
    """Change ``num'' to given base
    Upto base 36 is supported."""

    converted_string, modstring = "", ""
    currentnum = num
    if not 1 < base < 37:
        raise ValueError("base must be between 2 and 36")
    if not num:
        return '0'
    while currentnum:
        mod = currentnum % base
        currentnum = currentnum // base
        if (mod > 10):
            offset = 7
        else:
            offset = 0
        converted_string = chr(48 + mod + offset) + converted_string
    return converted_string

def nToBase10(num,base):

    #print (list(num))
    count = len(list(num))-1
    sum = 0
    for i in list(num):
        if (ord(i)>57):
            val = ord(i)-48-7
        else:
            val = ord(i)-48
        sum = sum + val*pow(base,count)
        #print(i,val,sum)
        count = count - 1
    return sum

def baseToBase(num,base1,base2):

    #print (list(num))
    count = len(list(num))-1
    sum = 0
    for i in list(num):
        if (ord(i)>57):
            val = ord(i)-48-7
        else:
            val = ord(i)-48
        sum = sum + val*pow(base1,count)
        #print(i,val,sum)
        count = count - 1
    return base10toN(sum,base2)


print("Q1:")
print(base10toN(86,2))
print(base10toN(763,2))
print(base10toN(3667,2))
print(base10toN(27538,2))
print(" ")
print("Q2:")
print(base10toN(86,8))
print(base10toN(763,8))
print(base10toN(3667,8))
print(base10toN(27538,8))
print(" ")
print("Q3:")
print(base10toN(86,16))
print(base10toN(763,16))
print(base10toN(3667,16))
print(base10toN(27538,16))
print(" ")
print("Q4:")
print(base10toN(86,5))
print(base10toN(763,5))
print(base10toN(3667,5))
print(base10toN(27538,5))
print(" ")
print("Q5:")
print(nToBase10("10111",2))
print(nToBase10("111010",2))
print(nToBase10("1010110",2))
print(nToBase10("10101101111",2))
print(" ")
print("Q6:")
print(nToBase10("3232",4))
print(nToBase10("2754",8))
print(nToBase10("100",16))
print(nToBase10("100",7))
print(" ")
print("Q7:")
print(baseToBase("E2",16,6))
print(baseToBase("1111001111",2,3))
print(baseToBase("BAD",14,20))
print(baseToBase("5345",6,16))
print(" ")
```
